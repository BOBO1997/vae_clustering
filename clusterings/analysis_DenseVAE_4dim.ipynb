```python
def visualize_zs(zs, labels, perplexity=30, learning_rate=200, n_iter=1000):
    plt.clf()
    plt.figure(figsize=(7,6))
    points = TSNE(n_components=2, random_state=0, perplexity=perplexity, learning_rate=learning_rate, n_iter=n_iter).fit_transform(zs)
    plt.scatter(points.transpose()[0], points.transpose()[1], s=20, c=labels)
    plt.colorbar()
    plt.show()
```

```python
def visualize_zsp(zs, perplexities, labels=None, learning_rate=200, n_iter=2000):
    plt.clf()
    fig = plt.figure(figsize=(len(perplexities) * 4, 4))
    for i, perplexity in enumerate(perplexities):
        points = TSNE(n_components=2, random_state=0, perplexity=perplexity, learning_rate=learning_rate, n_iter=n_iter).fit_transform(zs)
        ax = fig.add_subplot(1, len(perplexities), i + 1)
        ax.scatter(points.transpose()[0], points.transpose()[1], s=5) if labels is None else ax.scatter(points.transpose()[0], points.transpose()[1], s=5, c=labels)
    fig.show()
```

```python
perplexities = [5,25,45, 65,85,105,125,145]
visualize_zsp(hidden_mus[indices], perplexities, labels=s_labels)
```

## 最も近いものを取ってくる

```python
import heapq
class priority_queue:
    def __init__(self):
        self.key = lambda item: item[0] #アイテムの優先順位を指定
        self.index = 0
        self.data = []

    def push(self, item):
        heapq.heappush(self.data, (self.key(item), self.index, item))
        self.index += 1

    def pop(self):
        return heapq.heappop(self.data)[1]
```

```python
def get_nearest(pos, hidden_vecs, n = 10):
    t_vec = hidden_vecs[pos]
    pq = priority_queue()
    for i, vec in enumerate(hidden_vecs):
        dist = distance.euclidean(t_vec, vec)
        pq.push([dist, i])
    n_poses = []
    for i in range(n + 1):
        n_poses.append(pq.pop())
    return n_poses[1:]
```
